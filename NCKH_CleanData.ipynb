Chỉnh sửa định dạng ngày về chuẩn dd-mm-yyyy hh-mm

```python
import pandas as pd
from openpyxl import load_workbook
from datetime import datetime

def clean_and_format_excel(file_path, sheet_name):
    # Load the Excel file
    wb = load_workbook(file_path)
    ws = wb[sheet_name]

    # Iterate through column E and clean up data
    for row in ws.iter_rows(min_col=5, max_col=5, min_row=2):  # Assuming first row is header
        for cell in row:
            if cell.value:
                # Convert cell value to string and replace 'T' and 'Z'
                cleaned_value = str(cell.value).replace('T', ' ').replace('Z', '').replace('-', '/')
                if len(cleaned_value) > 19:  # Trim excess characters if longer than expected
                    cleaned_value = cleaned_value[:19]
                try:
                    # Convert to datetime and reformat
                    formatted_value = datetime.strptime(cleaned_value, "%Y/%m/%d %H:%M:%S")
                    cell.value = formatted_value.strftime("%d/%m/%Y %H:%M")
                except ValueError:
                    # If formatting fails, try alternative format
                    try:
                        formatted_value = datetime.strptime(cleaned_value, "%d/%m/%Y %H/%M")
                        cell.value = formatted_value.strftime("%d/%m/%Y %H:%M")
                    except ValueError:
                        # Keep the cleaned string if all else fails
                        cell.value = cleaned_value

    # Save the changes
    wb.save(file_path)

# Example usage
file_path = "Book1.xlsx"  # Replace with your file path
sheet_name = "Sheet1"  # Replace with your sheet name
clean_and_format_excel(file_path, sheet_name)
```

```python
import random
import string
from openpyxl import load_workbook

def generate_random_id(length=20):
    # Generate a random string of digits with the given length
    return ''.join(random.choices(string.digits, k=length))

def update_column_with_random_ids(file_path, sheet_name):
    # Load the Excel file
    wb = load_workbook(file_path)
    ws = wb[sheet_name]

    # Iterate through column A (column 1) and check if the value is a string
    for row in ws.iter_rows(min_col=1, max_col=1, min_row=2):  # Assuming first row is header
        for cell in row:
            if cell.value and isinstance(cell.value, str):  # Check if the value is a string
                # Replace with random 20 digit ID
                cell.value = generate_random_id()

    # Save the changes
    wb.save(file_path)

# Example usage
file_path = "Book1.xlsx"  # Replace with your file path
sheet_name = "Sheet1"  # Replace with your sheet name
update_column_with_random_ids(file_path, sheet_name)
```
